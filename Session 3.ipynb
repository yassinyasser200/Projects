```python
import pandas as pd
```

```python
data = pd.read_csv("laptop_data.csv")
```

```python
data.drop("Unnamed: 0" , axis =1 , inplace = True)
```

```python
data.TypeName=data.TypeName.str.replace("Ultrabook","Notebook").str.replace("Netbook","Notebook")
```

```python
data.Ram = data.Ram.str.replace("GB","")
```

```python
data.Ram = data.Ram.astype("int")
```

```python
data["IPS"] = data.ScreenResolution.apply( lambda x : 1 if "IPS" in x and "Touchscreen" not in x else 0)
```

```python
data["Touchscreen"] = data.ScreenResolution.apply( lambda x : 1 if "Touchscreen" in x and "IPS" not in x else 0)
```

```python
data["CPU_Name"] = data["Cpu"].str.split(" ").str[0:3].str.join(" ")
```

```python
data.ScreenResolution.str.split(" ").str[-1]
```

```python
data["Has_SSD"] = data.Memory.apply(lambda x : 1 if "SSD" in x else 0)
```

```python
data
```

```python
data["Has_HDD"] = data.Memory.apply(lambda x : 1 if "HDD" in x else 0)
```

```python
data["IsHyprid"] = data.Memory.apply(lambda x : 1 if "Hybrid" in x else 0)
```

```python
data
```

```python
data["Has_Flashstorge"] = data.Memory.apply(lambda x : 1 if "Flash Storage" in x else 0)
```

# Seaborn library

```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(12,8))
plt.xticks(rotation = 45)
plt.title("Total price per company" , fontsize = 18)
sns.barplot(data = data , x = "Company" , y= "Price" , color = "purple")
plt.show()
```

```python
plt.figure(figsize=(12,8))
plt.xticks(rotation = 45)
plt.title("Total Orders per company" , fontsize = 18)
sns.countplot(data = data , x = "Company" , color = "red"  )
plt.show()
```
