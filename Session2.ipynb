## Data Cleaning

```python
data.dtypes
```

### Fixing date datatype

```python
data.Date = pd.to_datetime(data.Date)
```

### fixing cost and price data types

```python
data["Unit Cost"] = data["Unit Cost"].str.replace("$","").str.replace(",","")
```

```python
data["Unit Cost"] = data["Unit Cost"].astype("int")
```

```python
data["Unit Cost"].dtype
```

```python
data["Unit Price"] = data["Unit Price"].str.replace("$","").str.replace(",","")
```

```python
data["Unit Price"] = data["Unit Price"].astype("int")
```

```python
data["Unit Price"].dtype
```

### Get the duplicated data 

```python
data[data.duplicated() == True]
```

```python
duplicated_data = data[data.duplicated() == True]
```

```python
duplicated_data
```

```python
data.drop_duplicates(inplace = True)
```

```python
data
```

## Data Extraction

### add total cost column

```python
data["Total Cost"] = data["Quantity"] * data["Unit Price"]
```

```python
data
```

```python
data["Profit"] = data["Total Cost"] - data["Unit Cost"]
```

```python
data
```

```python
data["test"] = 0
```

```python
data.drop(["test"] , axis = 1 , inplace = True)
```

```python
data["Month"] = data["Date"].dt.month_name()
```

```python
data
```

```python
data["Year"] = data["Date"].dt.year
```

```python
data
```

```python
data["Quarter"] = data.Date.dt.quarter
```

```python
data
```

```python
data['Profit Category'] = data['Profit'].apply(
    lambda x: 'Profit' if x > 0 else ('Lose' if x < 0 else 'constant')
)
```

```python
data.sample()
```

```python
def categorie (value):
    if(value > 0):
        return "positive profit"
    elif(value<0):
        return "Negative profit"
    else:
        return "Break Even"
```

```python
categorie(140)
```

```python
data["Categorized Profit"] = data.Profit.apply(categorie)
```

```python
data
```

```python
data["Month"] = data["Date"].dt.month_name().str[:3]
```

```python
data
```

```python
def age_class (value):
    if (17 <= value <= 35):
        return "Young Adult"
    elif( 36 <= value <= 59):
        return " Adult"
    else :
        return "Senior"
```

```python
data["Age Classification"] = data["Customer Age"].apply(age_class)
```
